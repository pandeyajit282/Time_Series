```python
def get_gold_price_data(start_date, end_date):
    gold_ticker = 'GC=F'
    
    gold_data = yf.download(gold_ticker, start=start_date, end=end_date)
    
    return gold_data

start_date = '2001-01-01'
end_date = '2023-01-01'

df = get_gold_price_data(start_date, end_date)
df
```

```python
df.drop(["Open", "High", "Low", "Adj Close", "Volume"], inplace=True,axis =1 )
type(df)
```

```python
def remove_outliers_iqr(data):
    # Step 1: Sort the data
    sorted_data = sorted(data)
    
    # Step 2: Calculate quartiles
    q1 = sorted_data[int(len(sorted_data) * 0.25)]
    q3 = sorted_data[int(len(sorted_data) * 0.75)]
    
    # Step 3: Calculate IQR
    iqr = q3 - q1
    
    # Step 4: Define outlier boundaries
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    
    # Step 5: Identify outliers
    outliers = [x for x in data if x < lower_bound or x > upper_bound]
    
    # Step 6: Remove outliers
    cleaned_data = [x for x in data if x >= lower_bound and x <= upper_bound]
    
    return cleaned_data, outliers

# Example usage:
df["Close"], outliers = remove_outliers_iqr(df["Close"])
```

```python
import pandas as pd

def normalize_data(data):

    min_vals = data.min()
    max_vals = data.max()
    
    # Normalize the data using Min-Max scaling
    normalized_data = (data - min_vals) / (max_vals - min_vals)
    
    return normalized_data

df.iloc[:,0] = normalize_data(df.iloc[:,0])
```

```python
# Perform seasonal decomposition
decomposition = seasonal_decompose(df["differenced_close"].dropna(), model='additive')  

# Plot the decomposed components
plt.figure(figsize=(12, 8))

plt.subplot(4, 1, 1)
plt.plot(decomposition.observed)
plt.title('Observed Data')

plt.subplot(4, 1, 2)
plt.plot(decomposition.trend)
plt.title('Trend')

plt.subplot(4, 1, 3)
plt.plot(decomposition.seasonal)
plt.title('Seasonality')

plt.subplot(4, 1, 4)
plt.plot(decomposition.resid)
plt.title('Residuals')

plt.tight_layout()
plt.show()
```

```python
ts = df1['Close']

# Define the rolling window size and forecast horizon
window_size = 50  
forecast_horizon = 30  
```

```python
# Initialize variables to keep track of training and testing indices
train_idx = 0
test_idx = train_idx + window_size + forecast_horizon
```

```python
# Iterate through the time series using the rolling window approach
while test_idx < len(ts):
    train_data = ts[train_idx:train_idx + window_size]
    test_data = ts[train_idx + window_size:test_idx]
    
    # Fit an ARIMA model on the training data
    order = (1,1,0)  
    model = ARIMA(train_data, order=order, seasonal_order=(4,1,3,12))
    fitted_model = model.fit()
    
    # Forecast future values using the fitted model
    forecast = fitted_model.forecast(steps=forecast_horizon)
    
    print("Forecasted values:")
    print(forecast)
    print("Actual test values:")
    print(test_data)
    
    # Move the window forward for the next iteration
    train_idx += forecast_horizon
    test_idx += forecast_horizon
```
